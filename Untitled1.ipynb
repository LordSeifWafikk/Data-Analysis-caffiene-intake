```python

import seaborn as sns
import matplotlib.pyplot as plt
import random

# Assuming 'df' is your DataFrame



# Define a function to generate random numeric values for GPA ranges
def map_gpa(gpa_range):
    if '-' in gpa_range:
        start, end = map(float, gpa_range.split(' - '))
        return round(random.uniform(start, end), 2)
    elif 'Above' in gpa_range:
        return round(random.uniform(3.6, 4.0), 2)
    elif 'Below' in gpa_range:
        return round(random.uniform(1.0, 1.99), 2)
    else:
        return float(gpa_range)


def map_caffeine_intake(intake_range):
    if 'Never' in intake_range:
        return 0
    elif '1-2 Times' in intake_range:
        return random.randint(1, 2)
    elif '3-4 Times' in intake_range:
        return random.randint(3, 4)
    elif '5-6 Times' in intake_range:
        return random.randint(5, 6)
    elif 'Daily' in intake_range:
        return 7  # Assuming 'Daily' is the highest range
    else:
        return 0


# Apply the function to map Cumulative GPA randomly within each range
df['What is your current accumulative GPA'] = df['What is your current accumulative GPA'].apply(map_gpa)
df['How often do you drink coffee per week'] = df['How often do you drink coffee per week'].apply(map_caffeine_intake)

# Check unique values in the column after mapping
print("\nUnique Values After Mapping:")
print("Cumulative GPA:", df['What is your current accumulative GPA'].unique())
print("Caffeine Intake:", df['How often do you drink coffee per week'].unique())

```

Some Calculations First

```python
import pandas as pd

# Assuming 'df' is your DataFrame

# Iterate through numeric columns and calculate mean, median, and mode
for column_name in df.select_dtypes(include='number').columns:
    mean_value = df[column_name].mean()
    median_value = df[column_name].median()
    mode_value = df[column_name].mode().values  # mode() returns a Series, so use values to get the array

    print(f"\nColumn: {column_name}")
    print(f"Mean: {mean_value}")
    print(f"Median: {median_value}")
    print(f"Mode: {mode_value}")
```

Scatterplot : Caffeine Intake vs Cumulative GPA

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame

# Set the style for the plot (optional)
sns.set(style="whitegrid")

# Create a scatter plot for Caffeine Intake and Cumulative GPA with switched axes
plt.figure(figsize=(10, 6))
sns.scatterplot(x='What is your current accumulative GPA', y='How often do you drink coffee per week', data=df, palette='viridis', s=30)
# Customize the plot
plt.title('Scatter Plot: Caffeine Intake vs Cumulative GPA')
plt.xlabel('Cumulative GPA')
plt.ylabel('Caffeine Intake')
plt.show()

```

```python
from scipy.stats import pearsonr

# Assuming 'df' is your DataFrame

# Calculate the Pearson correlation coefficient
correlation_coefficient, p_value = pearsonr( df['What is your current accumulative GPA'],df['How often do you drink coffee per week'])

# Display the correlation coefficient
print(f"Pearson Correlation Coefficient: {correlation_coefficient}")

```
